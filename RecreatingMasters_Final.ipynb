**Loading the data down below**:-

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
spx = pd.read_csv('https://raw.githubusercontent.com/Saneesh122/Data-Processing-Visualization-/main/SPX.csv', parse_dates = [0])
spx.head()
```

**Creating the chart of S&P 500 over the year down below**:-

```python
plt.plot(
        matplotlib.dates.date2num(spx['Date']), 
        spx['Adj Close'],
        color = 'green'
)

ax = plt.gca()
ax.xaxis.set_major_locator(   matplotlib.dates.YearLocator(10) )
ax.xaxis.set_major_formatter( matplotlib.dates.DateFormatter('%Y') )


plt.ylabel('S&P 500')
plt.yscale('log')
```

**Creating distributions of gains of S&P 500 through the year 1930 to 2020**:-

```python
spx_years = np.arange(1930, 2020)
spx_at_beginning_of_the_year = np.array([
    spx[spx.Date.dt.year == year]['Adj Close'].values[0]
    for 
    year in spx_years
])
```

```python
spx_annual_gains = np.array([
    spx_at_beginning_of_the_year[idx+1]/spx_at_beginning_of_the_year[idx]-1
    for 
    idx in range(len(spx_at_beginning_of_the_year)-1)
])
```

```python
plt.hist(spx_annual_gains, 10)
plt.title('Distribution of S&P 500 gains')
plt.show()
```

**Below visualtization gives us the idea of the flucation of S&P 500 from 1930 to 2020**:-

```python
plt.bar(spx_years[:-1], spx_annual_gains)
plt.ylabel('S&P 500 annual gains')
plt.title('Annual gains of S&P 500')
plt.show()
```

**S&P after the world war**:-

```python
spx_after_war = spx[ spx['Date'] > np.datetime64('1945') ]
```

```python
t     = matplotlib.dates.date2num(spx_after_war['Date'])
log_p = np.log10(spx_after_war['Adj Close'])

fit   = np.polyfit(t, log_p, deg = 1)

typical_daily_appreciation  = fit[0]
typical_annual_appreciation = 10**(365*typical_daily_appreciation) - 1.

print(f'Typical daily appreciation is {100*typical_daily_appreciation:.2f}%')
print(f'Typical annual appreciation is {100*typical_annual_appreciation:.2f}%')
```

```python
plt.plot( 
            matplotlib.dates.date2num(spx_after_war['Date']), 
            spx_after_war['Adj Close'],
            color = 'blue',
            label = 'S&P 500'
)
plt.plot(
            matplotlib.dates.date2num(spx_after_war['Date']), 
            10**(fit[0]*t + fit[1]),
            color = 'yellow',
            ls = '--',
            label = 'Fit'
)

plt.legend()

ax = plt.gca()
ax.xaxis.set_major_locator(matplotlib.dates.YearLocator(10))
ax.xaxis.set_major_formatter(matplotlib.dates.DateFormatter('%Y'))

plt.ylabel('S&P 500')
plt.yscale('log')
```
